**Cleaning Recipes.**

```python
all_recipes_df1 = pd.read_csv('recipes_data.csv', nrows=500000, usecols=['title', 'ingredients', 'directions', 'NER'])
all_recipes_df2 = pd.read_csv('recipes_data.csv', skiprows=range(1, 500001), nrows=500000, usecols=['title', 'ingredients', 'directions', 'NER'])
all_recipes_df3 = pd.read_csv('recipes_data.csv', skiprows=range(1, 1000001), nrows=500000, usecols=['title', 'ingredients', 'directions', 'NER'])
all_recipes_df4 = pd.read_csv('recipes_data.csv', skiprows=range(1, 1500001), nrows=500000, usecols=['title', 'ingredients', 'directions', 'NER'])
```

```python
def clean_ner(ner):
    return ast.literal_eval(ner)

all_recipes_df1['NER'] = all_recipes_df1['NER'].apply(clean_ner)
all_recipes_df2['NER'] = all_recipes_df2['NER'].apply(clean_ner)
all_recipes_df3['NER'] = all_recipes_df3['NER'].apply(clean_ner)
all_recipes_df4['NER'] = all_recipes_df4['NER'].apply(clean_ner)
```

```python
def valid_ingredients_percentage(ner_list):
    return sum([ingredient in valid_ingredients for ingredient in ner_list])

all_recipes_df1['valid_ingredient_percentage'] = all_recipes_df1['NER'].apply(valid_ingredients_percentage) / all_recipes_df1['NER'].apply(len) * 100
all_recipes_df2['valid_ingredient_percentage'] = all_recipes_df2['NER'].apply(valid_ingredients_percentage) / all_recipes_df2['NER'].apply(len) * 100
all_recipes_df3['valid_ingredient_percentage'] = all_recipes_df3['NER'].apply(valid_ingredients_percentage) / all_recipes_df3['NER'].apply(len) * 100
all_recipes_df4['valid_ingredient_percentage'] = all_recipes_df4['NER'].apply(valid_ingredients_percentage) / all_recipes_df4['NER'].apply(len) * 100
```

```python
valid_recipes_df1 = all_recipes_df1[all_recipes_df1['valid_ingredient_percentage'] >= 100]
valid_recipes_df2 = all_recipes_df2[all_recipes_df2['valid_ingredient_percentage'] >= 100]
valid_recipes_df3 = all_recipes_df3[all_recipes_df3['valid_ingredient_percentage'] >= 100]
valid_recipes_df4 = all_recipes_df4[all_recipes_df4['valid_ingredient_percentage'] >= 100]
```

```python
print(len(valid_recipes_df1['title'].unique()))
print(len(valid_recipes_df2['title'].unique()))
print(len(valid_recipes_df3['title'].unique()))
print(len(valid_recipes_df4['title'].unique()))
```

```python
recipes_df = pd.concat([valid_recipes_df1, valid_recipes_df2, valid_recipes_df3, valid_recipes_df4], ignore_index=True)
recipes_df = recipes_df.drop_duplicates(subset='title').reset_index().drop(columns='index')
recipes_df['carbon'] = 0
```
